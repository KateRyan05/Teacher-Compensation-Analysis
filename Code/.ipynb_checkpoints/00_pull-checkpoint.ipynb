# 00_pull.ipynb
**Purpose:** Load raw Stata data file, select the relevant variables, and write a cleaned CSV to `data/`.

| | |
|---|---|
| **Input** | `data/37941-0005-Data.dta` — ICPSR Wave 9 Stata file |
| **Output** | `data/raw_selected.csv` — subset of raw columns, no filtering |
| **Key functions** | `load_stata()`, `select_vars()` |

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#Paths
ROOT       = Path.cwd().parent
DATA_DIR   = ROOT / "Data"
OUTPUT_DIR = ROOT / "output"

INPUT_FILE  = DATA_DIR / "37941-0005-Data.dta"   # adjust per notebook
OUTPUT_FILE = DATA_DIR / "raw_selected.csv"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# Variable map
VARS = {
    'WF9_WORK_WAGE':              'hourly_wage',
    'WF9_CHAR_EDUC':              'education',
    'WF9_CHAR_GENDER':            'gender',
    'WF9_WORK_HRS_CAT':           'hours_cat',
    'WF9_CAREER_EXPERIENCE':      'experience',
    'WF9_CESD7_TOT':              'depression_score',
    'WF9_CHAR_HHINCOME':          'hh_income',
    'WF9_CHAR_YEAR_BORN':         'year_born',
    'WF9_REGION':                 'region',
    'WF9_WORK_FT':                'full_time_status',
    'WF9_WORK_ROLE':              'job_role',
    'WF9_WORK_MONTHS':            'months_per_year',
    'WF9_CHAR_RACE':              'race',
    'WF9_CHAR_HISP':              'hispanic_latino',
    'WF9_CHAR_COUNTRY_BORN':      'country_of_birth',
    'WF9_METH_PUBSCHFLAG':        'works_public_school',
    'WF9_CB_COMM_URBAN_DENSITY':  'urban_density',
    'WF9_PROFDEV_COURSE':         'college_for_pd',
    'WF9_PROFDEV_WRKSHP':         'workshop_for_pd',
    'WF9_A12_COLLENROLL':         'enrolled_in_college',
    'WF9_INTERACTIONS_TOT':       'interaction_quality_total',
    'WF9_E20_SRHEALTH':           'self_rated_health',
}
```

```python
#Functions
def select_vars(df: pd.DataFrame, var_map: dict) -> pd.DataFrame:
    subset = df[list(var_map.keys())].rename(columns=var_map)
    num_cols = subset.select_dtypes(include=[np.number]).columns
    subset[num_cols] = subset[num_cols].where(subset[num_cols] >= 0, np.nan)
    print(f"[select_vars] Selected shape: {subset.shape}")
    print(f"[select_vars] Columns: {subset.columns.tolist()}")
    return subset
```

```python
df_raw = pd.read_stata(INPUT_FILE)
#select analysis variables 
df_sel = select_vars(df_raw, VARS)

print("\n=== BEFORE WRITE ===")
print("Shape:")
print(df_sel.shape)
print("Columns:")
print(df_sel.columns.tolist())
print("First 3 rows:")
print(df_sel.head(3))

#save csv
df_sel.to_csv(OUTPUT_FILE, index=False)

df_check = pd.read_csv(OUTPUT_FILE)

print("\n=== AFTER WRITE ===")
print("Shape:", df_check.shape)
print("\nFirst 3 rows:")
print(df_check.head(3))

print(f"\nFile successfully saved to:\n{OUTPUT_FILE}")
```
