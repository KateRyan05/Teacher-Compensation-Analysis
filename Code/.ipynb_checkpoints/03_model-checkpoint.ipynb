# 03_model.ipynb
**Purpose:** Train OLS, Logistic Regression, Gradient Boosting, LightGBM, XGBoost, and CatBoost models on the cleaned data. Save train/test splits and model scores to `output/`.

| | |
|---|---|
| **Input** | `data/analysis_clean.csv` — output of `01_merge.ipynb` |
| **Output** | `output/model_scores.csv`, `output/fig_model_comparison.png`, `output/fig_ols_forest.png` |
| **Key functions** | `load_clean()`, `prep_model_data()`, `save_fig()`, `fit_ols()`, `fit_gb()`, `fit_lgbm()`, `fit_xgb()`, `fit_cat()`, `fit_logit()`, `plot_ols_forest()`, `plot_r2_comparison()`|

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import statsmodels.api as sm
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
import re
from pathlib import Path
```

```python
#Paths
ROOT       = Path.cwd().parent
DATA_DIR   = ROOT / "data"
OUTPUT_DIR = ROOT / "output"

INPUT_FILE  = DATA_DIR / "analysis_clean.csv"   # adjust per notebook
OUTPUT_FILE = DATA_DIR / "output"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

TARGET_COL = 'hourly_wage'

VARS = {
    'WF9_WORK_WAGE':              'hourly_wage',
    'WF9_CHAR_EDUC':              'education',
    'WF9_CHAR_GENDER':            'gender',
    'WF9_WORK_HRS_CAT':           'hours_cat',
    'WF9_CAREER_EXPERIENCE':      'experience',
    'WF9_CESD7_TOT':              'depression_score',
    'WF9_CHAR_HHINCOME':          'hh_income',
    'WF9_CHAR_YEAR_BORN':         'year_born',
    'WF9_REGION':                 'region',
    'WF9_WORK_FT':                'full_time_status',
    'WF9_WORK_ROLE':              'job_role',
    'WF9_WORK_MONTHS':            'months_per_year',
    'WF9_CHAR_RACE':              'race',
    'WF9_CHAR_HISP':              'hispanic_latino',
    'WF9_CHAR_COUNTRY_BORN':      'country_of_birth',
    'WF9_METH_PUBSCHFLAG':        'works_public_school',
    'WF9_CB_COMM_URBAN_DENSITY':  'urban_density',
    'WF9_PROFDEV_COURSE':         'college_for_pd',
    'WF9_PROFDEV_WRKSHP':         'workshop_for_pd',
    'WF9_A12_COLLENROLL':         'enrolled_in_college',
    'WF9_INTERACTIONS_TOT':       'interaction_quality_total',
    'WF9_E20_SRHEALTH':           'self_rated_health',
}

FEATURE_COLS = [v for v in VARS.values() if v != TARGET_COL]
RANDOM_STATE = 42
```

```python
#  Functions 
def load_clean(path: Path) -> pd.DataFrame:
    """Load the clean analysis CSV and drop any remaining all-NaN rows."""
    df = pd.read_csv(path)
    print(f"[load_clean] Shape: {df.shape}")
    return df

def prep_model_data(df: pd.DataFrame, feature_cols: list, target_col: str):
    model_df = df.copy()

    # Step 1: coerce target, drop non-numeric rows
    model_df[target_col] = pd.to_numeric(model_df[target_col], errors='coerce')
    before = len(model_df)
    model_df = model_df.dropna(subset=[target_col])
    print(f"[prep_model_data] Dropped {before - len(model_df)} rows with non-numeric target")

    X = model_df[feature_cols].copy()
    y = model_df[target_col]

    # Step 2: coerce numeric-looking object columns
    for col in X.columns:
        converted = pd.to_numeric(X[col], errors='coerce')
        if converted.notna().sum() > len(X) * 0.5:
            X[col] = converted

    # Step 3: label-encode remaining string/category columns
    for col in X.select_dtypes(include=['object', 'category']).columns:
        X[col] = pd.Categorical(X[col]).codes

    X = X.astype(float)

    # Step 4: drop any rows that still have NaN — before the split
    combined = X.copy()
    combined['__target__'] = y.values
    combined = combined.dropna()
    X = combined.drop(columns='__target__')
    y = combined['__target__']
    print(f"[prep_model_data] After dropping NaN rows: {X.shape}")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=RANDOM_STATE
    )
    print(f"[prep_model_data] Train: {X_train.shape}, Test: {X_test.shape}")
    return X, y, X_train, X_test, y_train, y_test

def save_fig(fig: plt.Figure, filename: str) -> None:
    """Save figure to output dir and close it."""
    dest = OUTPUT_DIR / filename
    fig.savefig(dest, bbox_inches='tight', dpi=150)
    plt.close(fig)
    print(f"[save_fig] Saved → {dest}")

def fit_ols(X_train, X_test, y_train, y_test, X, y):
    """Fit OLS and return model + metrics."""
    ols = LinearRegression()
    ols.fit(X_train, y_train)
    r2  = r2_score(y_test, ols.predict(X_test))
    cv  = cross_val_score(ols, X, y, cv=5, scoring='r2').mean()
    print(f"OLS         — Test R²: {r2:.4f} | CV R²: {cv:.4f}")
    return ols, r2, cv


def fit_gb(X_train, X_test, y_train, y_test, X, y):
    """Fit sklearn GradientBoostingRegressor and return model + metrics."""
    gb = GradientBoostingRegressor(n_estimators=300, learning_rate=0.05,
                                   max_depth=4, random_state=42)
    gb.fit(X_train, y_train)
    r2 = r2_score(y_test, gb.predict(X_test))
    cv = cross_val_score(gb, X, y, cv=5, scoring='r2').mean()
    print(f"GradBoost   — Test R²: {r2:.4f} | CV R²: {cv:.4f}")
    return gb, r2, cv


def fit_lgbm(X_train, X_test, y_train, y_test, X, y):
    """Fit LightGBM and return model + metrics."""
    lgbm = lgb.LGBMRegressor(n_estimators=500, learning_rate=0.3,
                              num_leaves=31, random_state=42, verbose=-1)
    lgbm.fit(X_train, y_train)
    r2 = r2_score(y_test, lgbm.predict(X_test))
    cv = cross_val_score(lgbm, X, y, cv=5, scoring='r2').mean()
    print(f"LightGBM    — Test R²: {r2:.4f} | CV R²: {cv:.4f}")
    return lgbm, r2, cv


def fit_xgb(X_train, X_test, y_train, y_test, X, y):
    """Fit XGBoost and return model + metrics."""
    xgbm = xgb.XGBRegressor(n_estimators=500, learning_rate=0.03,
                             max_depth=4, random_state=42, verbosity=0)
    xgbm.fit(X_train, y_train)
    r2 = r2_score(y_test, xgbm.predict(X_test))
    cv = cross_val_score(xgbm, X, y, cv=5, scoring='r2').mean()
    print(f"XGBoost     — Test R²: {r2:.4f} | CV R²: {cv:.4f}")
    return xgbm, r2, cv


def fit_cat(X_train, X_test, y_train, y_test, X, y):
    """Fit CatBoost and return model + metrics."""
    cat = CatBoostRegressor(iterations=500, learning_rate=0.05,
                            depth=4, random_seed=42, verbose=0)
    cat.fit(X_train, y_train)
    r2 = r2_score(y_test, cat.predict(X_test))
    cv = cross_val_score(cat, X, y, cv=5, scoring='r2').mean()
    print(f"CatBoost    — Test R²: {r2:.4f} | CV R²: {cv:.4f}")
    return cat, r2, cv


def fit_logit(df: pd.DataFrame, feature_cols: list, target_col: str):
    """Fit statsmodels Logit (above-median wage as outcome) and print summary."""
    df = df.copy()
    df[target_col] = pd.to_numeric(df[target_col], errors='coerce')
    df = df.dropna(subset=[target_col])

    median_wage = df[target_col].median()
    y_logit = (df[target_col] > median_wage).astype(int)

    X_logit = df[feature_cols].copy()

    # Encode ALL non-numeric columns, whether dtype is category or object
    for col in X_logit.select_dtypes(include=['category', 'object']).columns:
        X_logit[col] = pd.Categorical(X_logit[col]).codes

    X_logit = X_logit.astype(float)

    mask = X_logit.notna().all(axis=1)
    X_logit = X_logit[mask]
    y_logit = y_logit[mask]

    X_logit_sm = sm.add_constant(X_logit)
    logit_model = sm.Logit(y_logit.values, X_logit_sm.values).fit()
    print(logit_model.summary())
    return logit_model

def plot_ols_forest(ols, X_train, y_train) -> None:
    """Horizontal bar forest plot of OLS coefficients with approximate CIs.
    Derives feature names directly from X_train.columns to guarantee alignment."""
    feature_cols = X_train.columns.tolist()   # always in sync with ols.coef_

    coefs     = ols.coef_
    se_approx = (np.std(y_train - ols.predict(X_train))
                 / (np.sqrt(len(y_train)) * np.std(X_train.values, axis=0) + 1e-9))

    assert len(coefs) == len(feature_cols) == len(se_approx), (
        f"Length mismatch: coefs={len(coefs)}, "
        f"feature_cols={len(feature_cols)}, se={len(se_approx)}"
    )

    results = pd.DataFrame({
        'feat': feature_cols,
        'coef': coefs,
        'err':  se_approx * 1.96
    })
    results['label'] = results['feat'].str.replace('_', ' ').str.title()
    results = results.sort_values('coef')

    colors = ["#C44E52" if c < 0 else "#4C72B0" for c in results['coef']]

    fig, ax = plt.subplots(figsize=(9, 7))
    ax.barh(results['label'], results['coef'], xerr=results['err'],
            color=colors, alpha=0.8, edgecolor='white', height=0.6, capsize=4)
    ax.axvline(0, color='black', lw=0.9, ls='--')
    ax.set_xlabel("OLS Coefficient (effect on hourly wage $)")
    ax.set_title("OLS Forest Plot", fontsize=13, fontweight="bold")
    ax.grid(axis='x', alpha=0.3)

    blue_patch = mpatches.Patch(color="#4C72B0", alpha=0.8, label="Positive effect")
    red_patch  = mpatches.Patch(color="#C44E52", alpha=0.8, label="Negative effect")
    ax.legend(handles=[blue_patch, red_patch])

    plt.tight_layout()
    save_fig(fig, 'fig_ols_forest.png')

def plot_r2_comparison(model_names: list, r2_test: list, r2_cv: list) -> None:
    """Grouped bar chart comparing Test R² vs 5-Fold CV R² across all models."""
    colors = ["#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B2"]
    x = np.arange(len(model_names))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10, 6))
    bars1 = ax.bar(x - width/2, r2_test, width, label="Test R²",
                   color=colors, alpha=0.85, edgecolor='white')
    bars2 = ax.bar(x + width/2, r2_cv,   width, label="5-Fold CV R²",
                   color=colors, alpha=0.45, edgecolor='white', hatch='///')

    for bar in bars1:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.003,
                f"{bar.get_height():.3f}", ha='center', va='bottom',
                fontsize=10, fontweight='bold')
    for bar in bars2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.003,
                f"{bar.get_height():.3f}", ha='center', va='bottom',
                fontsize=10, color='grey')

    ax.set_ylabel("R²")
    ax.set_title("R² Comparison Across Models", fontsize=13, fontweight="bold")
    ax.set_xticks(x)
    ax.set_xticklabels(model_names)
    ax.legend()
    ax.set_ylim(0, max(r2_test + r2_cv) * 1.25)
    plt.tight_layout()
    save_fig(fig, 'fig_r2_comparison.png')
```

```python
# Load & prep
df = load_clean(INPUT_FILE)

print("\n── BEFORE MODELLING ──")
print(df[['hourly_wage', 'depression_score', 'experience',
          'hh_income', 'interaction_quality_total']].describe().round(2))

X, y, X_train, X_test, y_train, y_test = prep_model_data(df, FEATURE_COLS, TARGET_COL)
```

```python
print("df columns:", df.columns.tolist())
print("df shape:", df.shape)
print("df dtypes:\n", df.dtypes)
```

```python
# Fit all models
ols,  r2_ols,  cv_ols  = fit_ols(X_train, X_test, y_train, y_test, X, y)
gb,   r2_gb,   cv_gb   = fit_gb(X_train, X_test, y_train, y_test, X, y)
lgbm, r2_lgbm, cv_lgbm = fit_lgbm(X_train, X_test, y_train, y_test, X, y)
xgbm, r2_xgb,  cv_xgb  = fit_xgb(X_train, X_test, y_train, y_test, X, y)
cat,  r2_cat,  cv_cat  = fit_cat(X_train, X_test, y_train, y_test, X, y)
```

```python
# Logistic regression (above-median wage)
logit_model = fit_logit(df, FEATURE_COLS, TARGET_COL)
```
